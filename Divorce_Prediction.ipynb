### Data Cleaning

```python
import datetime
d['Date_of_marriage'] = pd.to_datetime(d['Date_of_marriage'])
d['Divorce_date'] = pd.to_datetime(d['Divorce_date'])
d['DOB_partner_man'] = pd.to_datetime(d['DOB_partner_man'])
d['DOB_partner_woman'] = pd.to_datetime(d['DOB_partner_woman'])
```

```python
d.info()
```

```python
d.head()
```

```python
d['Marriage_duration'] = d['Divorce_date'] - d['Date_of_marriage']
d['Marriage_duration'] = d['Marriage_duration']/np.timedelta64(1,'D')
d.head()
```

```python
d.dropna(subset = ['DOB_partner_man','DOB_partner_woman'], inplace = True)
```

```python
d.isnull().sum()
```

## Date Time Conversion

```python
d['mday'] = d['DOB_partner_man'].dt.day
d['mmonth'] = d['DOB_partner_man'].dt.month
d['wday'] = d['DOB_partner_woman'].dt.day
d['wmonth'] = d['DOB_partner_woman'].dt.month
```

```python
d.head()
```

### Zodiac Compatability

```python
d['mz'] = d['mday']+(100*d['mmonth'])
d['wz'] = d['wday']+(100*d['wmonth'])
```

```python
d.drop(['Date_of_marriage','Divorce_date','DOB_partner_man','DOB_partner_woman','mday','wday','mmonth','wmonth'], axis = 1, inplace = True)
```

```python
d.head()
```

```python
def zodiac(x):
  if x >= 1222 or x <= 120 :
    return 'Capricorn'
  elif x >= 121 and x <= 219:
    return 'Aquarius'
  elif x >= 220 and x <= 320:
    return 'Pisces'
  elif x >= 321 and x <= 420:
    return 'Aries'
  elif x >= 421 and x <= 521:
    return 'Taurus'
  elif x >= 522 and x <= 621:
    return 'Gemini'
  elif x >= 622 and x <= 722:
    return 'Cancer'
  elif x >= 723 and x <= 822:
    return 'Leo'
  elif x >= 823 and x <= 922:
    return 'Virgo'
  elif x >= 923 and x <= 1022:
    return 'Libra'
  elif x >= 1023 and x <= 1122:
    return 'Scorpio'
  elif x >= 1123 and x <= 1221:
    return 'Sagittarius'
  else:
    return 'Capricon' 
```

```python
d['man_zodiac'] = d['mz'].apply(zodiac)
d['woman_zodiac'] = d['wz'].apply(zodiac)
d.head()
```

```python
d.drop(['mz','wz'], axis = 1, inplace = True)
```

```python
d['Zodiac_combination'] = d['man_zodiac'] + d['woman_zodiac']
d.head()
```

## Dimensionality Reduction

```python
from sklearn.decomposition import PCA
dfq = df.drop('Compatibility_rate', axis = 1)
pca = PCA(n_components = 2)
dft = pca.fit_transform(dfq)
dft.shape
```

```python
dft = pd.DataFrame(dft, columns = ['f1','f2'])
```

### Clustering

```python
plt.figure(figsize = (15,9))
dft.plot(kind = 'scatter',x= 'f1', y = 'f2',s = df['Num_Children']*40, c = df['Marriage_duration'], alpha = 0.7, cmap = 'inferno', colorbar = True, figsize = (15,9));
```

```python
plt.figure(figsize = (15,9))
dft.plot(kind = 'scatter',x= 'f1', y = 'f2',s = df['Num_Children']*40, c = df['Compatibility_rate'], alpha = 0.7, cmap = 'inferno', colorbar = True, figsize = (15,9));
```

```python
from sklearn.cluster import KMeans

w = []
for i in range(1,11):
    k = KMeans(n_clusters = i, init = 'k-means++', random_state = 9)
    k.fit(dft)
    
    w.append(k.inertia_)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

sns.set()
plt.plot(range(1,11),w)
plt.xlabel('No. of Clusters')
plt.ylabel('W')
plt.title('Elbow Point Graph');
```

```python
km = KMeans(n_clusters = 2, init = 'k-means++', random_state = 22)

y = km.fit_predict(dft)
plt.figure(figsize = (15,9))
dft.plot(kind = 'scatter',x= 'f1', y = 'f2',s = df['Num_Children']*40, c = df['Marriage_duration'], alpha = 0.7, cmap = 'jet', colorbar = True, figsize = (15,9));
```

```python
dft['target'] = df['Marriage_duration']
dft
```

```python
q = dft.quantile(q = [0.33,0.66])
q.to_dict()
```

```python
def dv(x,p,d):
    if x <= d[p][0.33]:
        return 1
    elif x <= d[p][0.66]:
        return 2
    else:
        return 3
```

```python
dft['divorce'] = dft['target'].apply(dv, args = ('target', q))
dft.head()
```

```python
dft['divorce'].value_counts()
```

```python
plt.figure(figsize = (15,9))
dft.plot(kind = 'scatter',x= 'f1', y = 'f2',s = df['Num_Children']*40, c = 'divorce', alpha = 0.7, cmap = 'jet', colorbar = True, figsize = (15,9));
```

```python
plt.plot(range(len(names)), lso_co)
plt.xticks(range(len(names)),names, rotation = 90)
plt.title('feature importance')
plt.ylabel('coefficients')
```
